```python
s1 = int(input("Eglish subject 1: "))
s2 = int(input("Math 2: "))
s3 = int(input("physics 3: "))
s4 = int(input("chemistry 4: "))
s5 = int(input("urdu 5: "))

total = s1 + s2 + s3 + s4 + s5
percentage = total / 5

print("Total Marks:", total)
print("Percentage:", percentage)

if percentage >= 80:
    print("Grade: A")
elif percentage >= 60:
    print("Grade: B")
elif percentage >= 40:
    print("Grade: C")
else:
    print("Grade: Fail")
    
```
